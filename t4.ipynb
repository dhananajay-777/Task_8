```python
import speech_recognition as sr
import wave
import sys
import pyaudio
def speech_to_text():
    r = sr.Recognizer()
    with sr.Microphone() as s1:
        r.adjust_for_ambient_noise(s1,duration=0.1)
        print("tell your answer:")
        audio = r.listen(s1,timeout=1000,phrase_time_limit=10,)
        try:
            text = r.recognize_google(audio,language = "en-IN")
            return text
        except:
            return 0
        
```

```python
import random
import datetime
from time import sleep
# Global List Declaration
name = []
phno = []
add = []
checkin = []
checkout = []
room = []
price = []
rc = []
p = []
roomno = []
custid = []

  
# Global Variable Declaration
  
i = 0
  
```

```python
# Booking function
def Booking():
     
        # used global keyword to
        # use global variable 'i'
        global i
        print(" BOOKING ROOMS")
        print(" Please answer the saying out the answer aloud.")
         
        while 1:
            print("Name: ")
            n = str(speech_to_text())
            sleep(1)
            print("Phone No.: ")
            p1 = str(speech_to_text())
            sleep(1)
            print("Address: ")
            a = str(speech_to_text())
             
            # checks if any field is not empty
            if n!="" and p1!="" and a!="":
                name.append(n)
                add.append(a)
                phno.append(p1)
                break
                 
            else:
                 print("\tName, Phone no. & Address cannot be empty..!!")
        sleep(1)
        print("Check-In: (dd/mm/yyyy)")
        cii=str(speech_to_text())
        checkin.append(cii)
        sleep(1)
        print("Check-Out: (dd/mm/yyyy)")
        coo=str(speech_to_text())
        checkout.append(coo)

        sleep(2)
        print("----SELECT ROOM TYPE----")
        print(" Tell 1 for Standard Non-AC")
        print(" Tell 2  Standard AC")
        print(" Tell 3 3-Bed Non-AC")
        print(" Tell 4 3-Bed AC")
        print(("\t\tTell 0 for Room Prices"))
         
        ch=int(speech_to_text())
         
        # if-conditions to display allotted room
        # type and it's price
        if ch==0:
            print(" 1. Standard Non-AC - Rs. 3500")
            print(" 2. Standard AC - Rs. 4000")
            print(" 3. 3-Bed Non-AC - Rs. 4500")
            print(" 4. 3-Bed AC - Rs. 5000")

            ch=int(speech_to_text())
        if ch==1:
            room.append('Standard Non-AC')
            print("Room Type- Standard Non-AC") 
            price.append(3500)
            print("Price- 3500")
        elif ch==2:
            room.append('Standard AC')
            print("Room Type- Standard AC")
            price.append(4000)
            print("Price- 4000")
        elif ch==3:
            room.append('3-Bed Non-AC')
            print("Room Type- 3-Bed Non-AC")
            price.append(4500)
            print("Price- 4500")
        elif ch==4:
            room.append('3-Bed AC')
            print("Room Type- 3-Bed AC")
            price.append(5000)
            print("Price- 5000")
        else:
            print(" Wrong choice..!!")
  
 
        # randomly generating room no. and customer
        # id for customer
        rn = random.randrange(40)+300
        cid = random.randrange(40)+10
         
         
        # checks if allotted room no. & customer
        # id already not allotted
        while rn in roomno or cid in custid:
            rn = random.randrange(60)+300
            cid = random.randrange(60)+10
             
        rc.append(0)
        p.append(0)
               
        print("")
        print("\t\t\t***ROOM BOOKED SUCCESSFULLY***\n")
        print("Room No. - ",rn)
        print("Customer Id - ",cid)
        roomno.append(rn)
        custid.append(cid)
        i=i+1
        exit()
 
```

```python
Booking()
```
